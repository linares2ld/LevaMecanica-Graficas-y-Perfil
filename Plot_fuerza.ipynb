```python
import numpy as np
import matplotlib.pyplot as plt
import camdesign as cd
```

# Gr√°fica fuerza - tiempo 

```python
x = np.linspace(0,0.5, 361)

y1 = cd.Cicloidal(113, x[0:57], 0.0789, "ascenso")
y2 = cd.Cicloidal(33,x[0:68],0.0942, "descenso")
y3 = cd.Cicloidal(28,x[0:75],0.1039, "ascenso")
y4 = cd.Cicloidal(108,x[0:60],0.0831, "descenso")
y5 = np.zeros(101)
y = []
y = np.concatenate([y1, y2 + 80, y3 + 80, y4, y5])

plt.figure(figsize=(10, 4))
plt.plot(x, y, label = "Movimiento Cicloidal")

plt.grid(True)
plt.xlabel('s')
plt.axhline(0, color='black', linewidth=0.5)
plt.xticks(np.linspace(0,0.5,11))
plt.ylabel('lb')

plt.title('Fuerza - tiempo')
plt.legend()
plt.show()
```
